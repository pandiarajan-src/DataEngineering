```python
### Create a phone book as a dictionary with names and numbers

phone_book = {'name':['Pandi', 'Bhuvana', 'Kaavya', 'Subhanu'], 'mobile_numbers': ['12345', "54321", "67890", "09876",]}
print(phone_book)

# Sort a dictionary by key and value
sorted_phone_book = sorted(phone_book)
sorted_phone_book
```

```python
# Build a set of unique words from a long input string
import re
email = "hi how are you? you won a lottery with the price money of lot, are you happy? money is money at the end"
cleaned_email_line = re.sub(r'[^a-zA-Z\s]', '', email)
email_unique_word_token = set(cleaned_email_line.split(" "))
email_unique_word_token
```

```python
# Compare arrays of numbers using set intersections and differences

set1 = set([1, 2, 4, 5, 6, 7, 7, 8, 9, 10, 11])
set2 = set([11, 22, 44, 55, 6, 7, 88, 99, 10])

res  = set1.intersection(set2)
print(f"result of set1 and set2 intersection is {res}")

res  = set1.difference(set2)
print(f"result of set1 and set2 difference is {res}")

res  = set1.union(set2)
print(f"result of set1 and set2 union is {res}")
```

```python
# Sort a dictionary by key and by value

phone_book = {'Pandi': "12345", 'Bhuvana': "67890", 'Kaavya': "54321", 'Subhanu': '09876'}

# Sort by keys as default
sorted_phone_book_by_name = dict(sorted(phone_book.items()))
print(f"sorted phone book by keys (by name) is {sorted_phone_book_by_name}")

# Sort by values
sorted_phone_book_by_number = dict(sorted(phone_book.items(), key=lambda x: x[1]))
print(f"sorted phone book by values (by number) is {sorted_phone_book_by_number}")
```

```python
# Create histograms of auto generated numbers

import numpy as np
import matplotlib.pyplot as plt

mapping = {1: "one", 2: "two", 3: "three", 4: "four", 5: "five", 6: "six", 7: "seven", 8: "eight", 9: "nine", 10: "ten"}

arr = np.random.randint(1, 10, 1000)
hist = {}

for x in arr:
    if mapping[x] not in hist:
        hist[mapping[x]] = 0
    else:
        hist[mapping[x]] += 1

print(hist)

plt.title("Histograms of auto generated numbers")
plt.xlabel("Numbers")
plt.ylabel("Count")
plt.bar(hist.keys(), hist.values(), color='b')
plt.show()
```

```python
# Create histograms of word counts using dictionaries

import matplotlib.pyplot as plt
import re

email = "hi how are you? you won a lottery with the price money of lot, are you happy? money is money at the end"
cleaned_email_line = re.sub(r'[^a-zA-Z\s]', '', email)

word_count = {}

for word in cleaned_email_line.split(" "):
    if word not in word_count:
        word_count[word] = 0
    else:
        word_count[word] += 1

print(f"given email: {email}")
print(f"unique words: {cleaned_email_line}")
print(f"word count: {word_count}")

plt.title("Histograms of word counts")
plt.xlabel("Words")
plt.ylabel("Count")
plt.bar(word_count.keys(), word_count.values(), color='b')
plt.show()
```
